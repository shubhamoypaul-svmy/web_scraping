```python
# Perform further parsing and extraction using BeautifulSoup as usual
from bs4 import BeautifulSoup
import requests

# URL of the webpage with the search bar
url = "https://www.moneycontrol.com/news/tags/nifty.html/news/"  # Replace with the actual URL
response = requests.get(url)

#class="YMlKec fxKbKc"
soup = BeautifulSoup(response.content, "html.parser")
lines = soup.find_all("li", class_="clearfix")
```

```python
links = []
Title = []
start = []
date = []
content = []
for line in lines:
  links.append(line.find("a").get('href'))
  Title.append(line.find("a").get('title'))
  start.append(line.find("p").text)
  date.append(line.find("span").text)
  response = requests.get(line.find("a").get('href'))
  html_content = response.content

  soup = BeautifulSoup(html_content, 'html.parser')

  # Find all paragraphs on the page
  paragraphs = soup.find('div', id='contentdata')
  if paragraphs is None:
    content.append('None')
  else:
    content.append(paragraphs.text)
```

```python
import pandas as pd
df = pd.DataFrame(
  columns = ['Title', 'Date', 'Link', 'Start', 'Content']
)
df['Title'] = Title
df['Content'] = content
df['Date'] = date
df['Link'] = links
df['Start'] = start
df.to_csv('news.csv', index=False)
```
